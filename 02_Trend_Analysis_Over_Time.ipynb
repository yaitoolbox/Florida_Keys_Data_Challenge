# Trend Analysis of Multiple Coral Species (3 Per Row)

**Purpose:** This section visualizes the trends of multiple coral species on a single figure, arranged in subplots with 3 species per row. This allows for easier comparison of trends across different species.

**Inputs:**

* `processed_stony_coral.csv` (from "temp/" directory)

**Outputs:**

* A figure containing subplots, where each subplot shows the trend for one coral species.

**Analysis Steps:**

1.  **(Previous Steps 1 & 2 - Data Loading and Inspection - Assumed to be completed)**
2.  **Define Coral Species:** Create a list of the coral species you want to analyze.
3.  **Create Subplots:**
    * Calculate the number of rows needed for the subplots (based on the number of species and 3 columns per row).
    * Create a figure and a grid of subplots.
4.  **Plot Trends:**
    * Iterate through the list of coral species.
    * For each species, group the data by 'year' and calculate the mean cover.
    * Plot the trend in the corresponding subplot.
5.  **Customize Layout:** Adjust the layout of the subplots to prevent overlapping.

**Notes:**

* This section assumes that Steps 1 and 2 have been executed.
* The number of rows in the plot is dynamically calculated to accommodate any number of species.
* You can customize the figure size and other plot parameters as needed.

```python
import pandas as pd
import os
import matplotlib.pyplot as plt

# Step 3: Analyze Multiple Coral Species (3 Per Row)

# Define the coral species to analyze
coral_species_to_plot = [
    'Acropora_cervicornis',
    'Porites_astreoides',
    'Montastraea_cavernosa',
    'Diploria_labyrinthiformis',
    'Millepora_alcicornis',
    'Agaricia_fragilis',
    'Colpophyllia_natans',
    'Dendrogyra_cylindrus',
    'Siderastrea_siderea'  # Add more species as needed
]

# Define colors for each species
colors = [
    'blue', 'green', 'red', 'purple', 'orange',
    'brown', 'pink', 'cyan', 'magenta', 'lime',
    'teal', 'olive', 'navy', 'maroon'
]

# Calculate the number of rows and columns for the subplots
num_species = len(coral_species_to_plot)
num_cols = 3
num_rows = (num_species + num_cols - 1) // num_cols

# Create the figure and subplots
fig, axes = plt.subplots(nrows=num_rows, ncols=num_cols, figsize=(15, 5 * num_rows))
fig.tight_layout(pad=5.0)

# Plot the trends for each coral species
for i, species in enumerate(coral_species_to_plot):
    row_index = i // num_cols
    col_index = i % num_cols

    try:
        mean_cover_by_year = df_stony_coral.groupby('year')[species].mean()
    except KeyError:
        print(f"Error: Column '{species}' not found. Skipping {species}.")
        continue

    # Select the correct subplot axes
    if num_rows > 1:
        ax = axes[row_index, col_index]
    elif num_cols > 1:
        ax = axes[col_index]
    else:
        ax = axes

    ax.plot(mean_cover_by_year.index, mean_cover_by_year.values, marker='o', color=colors[i % len(colors)])
    ax.set_xlabel("Year")
    ax.set_ylabel("Mean Cover")
    ax.set_title(species)
    ax.grid(True)

# Remove any empty subplots
if num_species % num_cols != 0 and num_rows > 1:
    for j in range(num_species % num_cols, num_cols):
        fig.delaxes(axes[num_rows - 1, j])
elif num_species % num_cols != 0 and num_cols > 1 and num_rows <= 1:
    for j in range(num_species % num_cols, num_cols):
      fig.delaxes(axes[j])

plt.show()
```

# <span style="color:blue;">Estimated Coral Cover Trends (2024-2026)</span>
#
**Purpose:** This section extends the coral cover trend analysis to provide *estimated* coral cover values for 2024, 2025, and 2026, based on a linear regression model fitted to historical data.

**Inputs:**

* `processed_stony_coral.csv` (from "temp/" directory)

**Outputs:**

* Visualizations showing historical coral cover trends along with estimated values for 2024-2026.
* DataFrames containing the estimated coral cover values.

**Analysis Steps:**

1.  **(Previous Steps 1 & 2 - Data Loading and Inspection - Assumed to be completed)**
2.  **Define Coral Species:** Create a list of the coral species to analyze.
3.  **Fit Linear Regression Model:**
    * For each coral species:
        * Prepare the data for regression (year as the independent variable, mean coral cover as the dependent variable).
        * Fit a linear regression model to the historical data.
4.  **Generate Estimates:**
    * Use the fitted model to predict coral cover for 2024, 2025, and 2026.
5.  **Visualize Results:**
    * Plot the historical coral cover data.
    * Plot the estimated coral cover values for 2024-2026.
    * Clearly distinguish between historical data and estimates in the plot.

**Notes:**

* This analysis uses a simple linear regression model for estimation. The accuracy of the estimates is subject to the limitations of this model and the nature of the data.
* The `scikit-learn` library is used for linear regression.

```python
import pandas as pd
import os
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import numpy as np

# Step 3: Estimate Coral Cover Trends (2024-2026)

# Define the coral species to analyze
coral_species_to_plot = [
    'Acropora_cervicornis',
    'Porites_astreoides',
    'Montastraea_cavernosa',
    'Diploria_labyrinthiformis',
    'Millepora_alcicornis',
    'Agaricia_fragilis',
    'Colpophyllia_natans',
    'Dendrogyra_cylindrus',
    'Siderastrea_siderea'  # Add more species as needed
]

# Years to predict
years_to_predict = [2024, 2025, 2026]

# Calculate the number of rows and columns for the subplots
num_species = len(coral_species_to_plot)
num_cols = 3
num_rows = (num_species + num_cols - 1) // num_cols

# Create the figure and subplots
fig, axes = plt.subplots(nrows=num_rows, ncols=num_cols, figsize=(15, 5 * num_rows))
fig.tight_layout(pad=5.0)

# Store estimated data
estimated_data = {}

# Plot the trends and estimates for each coral species
for i, species in enumerate(coral_species_to_plot):
    row_index = i // num_cols
    col_index = i % num_cols

    try:
        mean_cover_by_year = df_stony_coral.groupby('year')[species].mean().reset_index()
    except KeyError:
        print(f"Error: Column '{species}' not found. Skipping {species}.")
        continue

    # Prepare data for regression
    X = mean_cover_by_year['year'].values.reshape(-1, 1)  # Year
    y = mean_cover_by_year[species].values  # Coral cover

    # Fit linear regression model
    model = LinearRegression()
    model.fit(X, y)

    # Generate estimates for future years
    future_X = np.array(years_to_predict).reshape(-1, 1)
    future_y = model.predict(future_X)

    # Store estimated data
    estimated_data[species] = pd.DataFrame({'year': years_to_predict, 'estimated_cover': future_y})

    # Select the correct subplot axes
    if num_rows > 1:
        ax = axes[row_index, col_index]
    elif num_cols > 1:
        ax = axes[col_index]
    else:
        ax = axes

    # Plot historical data
    ax.plot(mean_cover_by_year['year'], mean_cover_by_year[species], marker='o', label='Historical')
    # Plot estimated data
    ax.plot(years_to_predict, future_y, marker='x', linestyle='--', color='red', label='Estimated')

    ax.set_xlabel("Year")
    ax.set_ylabel("Mean Cover")
    ax.set_title(species)
    ax.grid(True)
    ax.legend()  # Show legend

# Remove any empty subplots
if num_species % num_cols != 0 and num_rows > 1:
    for j in range(num_species % num_cols, num_cols):
        fig.delaxes(axes[num_rows - 1, j])
elif num_species % num_cols != 0 and num_cols > 1 and num_rows <= 1:
    for j in range(num_species % num_cols, num_cols):
        fig.delaxes(axes[j])

plt.show()

# Print estimated data (optional)
for species, data in estimated_data.items():
    print(f"\n--- Estimated Coral Cover for {species} ---")
    print(data)
```
